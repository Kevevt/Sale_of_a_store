### Escribe un programa que lea tres números y los muestre en orden descendente.

```python
num1 = int(input("Ingrese número 1: "))
num2 = int(input("Ingrese número 2: "))
num3 = int(input("Ingrese número 3: "))

if num2 < num1 > num3:
    if num2 > num3:
        print(f'El orden es así: {num1}, {num2}, {num3}')
    else:
        print(f'El orden es así: {num1}, {num3}, {num2}')
elif num1 < num2 > num3:
    if num3>num1:
        print(f'El orden es así: {num2}, {num3}, {num1}')
    else:
        print(f'El orden es así: {num2}, {num1}, {num3}')
elif num1>num2:
    print(f'El orden es así: {num3}, {num1}, {num2}')
else:
    print(f'El orden es así: {num3}, {num2}, {num1}')
```

### 2 - Escribe un programa para calcular cuántos días tomará que la colonia de una bacteria A supere o iguale a la colonia de una bacteria B, basado en tasas de crecimiento del 3% y 1.5%, respectivamente. Supón que la colonia A comienza con 4 elementos y B con 10.

```python
a = 4
b = 10
dia = 0

while a < b:
    dia += 1
    a = (a)*0.03 + a
    b = (b)*0.015 +b
    # a (0.03 +1) = a(1.03)
print(f'tomó esta cantidad de días: {dia}')
print(a, b)
```

```python
a = 4.0
b = 10.0
dias = 0

while a < b:          # sigue hasta que A >= B
    a *= 1.03         # +3%
    b *= 1.015        # +1.5%
    dias += 1

print(f"Días: {dias}")
```

```python
# Número inicial de bacterias
colonia_a = 4
colonia_b = 10

# Tasas de crecimiento de las colonias
tasa_a = 0.03
tasa_b = 0.015

# Contador de días
días = 0

# La condición que termina el bucle es cuando
# la colonia A supera a la colonia B
while colonia_a <= colonia_b:
  # Usamos un operador de asignación con multiplicación
  colonia_a *= 1 + tasa_a
  colonia_b *= 1 + tasa_b
  # Contamos los días en cada iteración
  días += 1

# Resultado final
print(f'Necesitará {días} días para que la colonia A supere a la colonia B.')
```
