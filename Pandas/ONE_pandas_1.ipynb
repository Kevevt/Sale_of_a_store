### Ejemplos sencillos:

```python
import pandas as pd

df = pd.DataFrame({
   'Animal': ['Perro', 'Gato', 'Elefante', 'Perro', 'Gato', 'Elefante'],
   'Color': ['Negro', 'Blanco', 'Gris', 'Marrón', 'Negro', 'Marrón'],
   'Cantidad': [2, 3, 1, 4, 2, 2]
})
df
```

```python
dff = df.groupby('Animal')['Cantidad'].sum(numeric_only=True)
dff.plot(kind='bar',color='green')
```

```python
df.head()
```

```python
df.groupby(['Animal'])[['Cantidad', 'Color']].sum()
```

```python
# Así es como le dices qué hacer con cada columna
df.groupby('Animal').agg({
    'Cantidad': 'sum',
    'Color': list
})
```
