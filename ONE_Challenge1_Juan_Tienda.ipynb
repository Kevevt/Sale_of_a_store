```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

## **0. Unión, verificación y limpieza de datos**

```python
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
```

```python
display(tienda1.head())
display(tienda2.head())
display(tienda3.head())
display(tienda4.head())
```

Unir los 4 dataFrame en uno solo

```python
tiendas_totales = pd.concat([tienda1,tienda2,tienda3,tienda4])
tiendas_totales.sample(5)
```

```python
tiendas_totales.dtypes
```

## **1. Análisis de factuación**

### **Visualización cruda**

```python
tiendas_graf_factura =tiendas_totales.groupby('Tienda')[['Precio']].sum().round(2).sort_values('Precio')
tiendas_graf_factura.plot(kind='barh')
```

### **Visualización con objetivo**

```python
import matplotlib.pyplot as plt

df_plot = tiendas_graf_factura.copy()
df_plot['Precio'] = df_plot['Precio'] / 1_000_000

df_plot = df_plot.sort_values('Precio', ascending=False)

colores = ['#C0C0C0'] * (len(df_plot) - 1) + ['#D62728']

fig, ax = plt.subplots(figsize=(12, 6))

df_plot.plot(kind='barh', y='Precio', ax=ax, color=colores, width=0.65, legend=False)

ax.set_title('Análisis de Facturación: La Tienda 4 muestra el desempeño más bajo',
             loc='left', fontsize=16, fontweight='bold', color='#333333', pad=20)

ax.set_xlabel('Facturación Total (Millones de Pesos)', fontsize=11, color='#555555')
ax.set_ylabel('') 

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#DDDDDD') 
ax.spines['bottom'].set_color('#DDDDDD')

ax.tick_params(axis='y', length=0)

for container in ax.containers:
    
    ax.bar_label(container, fmt='%.0f M', padding=5, fontsize=10, color='#333333')

plt.tight_layout()

plt.show()
```

## **2. Ventas por categoría**

```python
tiendas_graf_cat =tiendas_totales.groupby(['Tienda', 'Categoría del Producto'])[['Precio']].sum().round(2)
tiendas_graf_cat.sample(10)
```

### Visualización cruda

```python
# 1. Des-apilar (Convertir categorías en columnas)
tabla_grafico_2 = tiendas_graf_cat.unstack()

# 2. Ver la tabla transformada
display(tabla_grafico_2)

# 3. ¡Graficar directo!
# Al ser una matriz, pandas hace barras agrupadas automáticamente
tabla_grafico_2.plot(kind='bar', figsize=(12,6))
```

### **Visualización con objetivo**

```python
# PREPARACIÓN Y FILTRADO
df_cat_plot = tabla_grafico_2.copy()
df_cat_plot.columns = df_cat_plot.columns.droplevel(0)
df_cat_plot = df_cat_plot / 1_000_000

# --- 1. Selección del Top 3
total_por_categoria = df_cat_plot.sum().sort_values(ascending=False)
top_3_nombres = total_por_categoria.head(3).index
df_final = df_cat_plot[top_3_nombres]

# --- 1. DATOS ---
# df_final: índices = tiendas, columnas = categorías (en millones de MXN)
colores_estilo = ['#003366', '#5DADE2', '#999999']

fig, ax = plt.subplots(figsize=(12, 7))

# --- 2. BARRAS ---
df_final.plot(
    kind='bar',
    ax=ax,
    width=0.75,
    color=colores_estilo,
    edgecolor='none'        # sin borde blanco entre barras
)

# --- 3. GRID Y EJES ---
# Grid horizontal suave
ax.grid(axis='y', linestyle='--', linewidth=0.5, color='gray', alpha=0.3)
ax.set_axisbelow(True)

# Título
ax.set_title(
    'Ventas por categoría (millones de Pesos): Top 3 productos',
    loc='left',
    fontsize=16,
    fontweight='bold',
    color='#333333',
    pad=30
)

# Ejes
#ax.set_ylabel('Ventas (Millones de Pesos)', fontsize=12, color='#333333', labelpad=10)
ax.set_xlabel('')

plt.xticks(rotation=0, fontsize=13)
plt.yticks(fontsize=10)

# --- 4. TICKS Y SPINES ---
# Sin rayitas en el eje X
ax.tick_params(axis='x', length=0, bottom=False)

# Eje Y: sin rayitas, solo números + grid
ax.tick_params(axis='y', length=0)

# Spines (bordes del gráfico)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)    # si quieres un eje Y marcado, pon True aquí
ax.spines['bottom'].set_color("#F0F0F0")

# --- 5. LEYENDA ---
ax.legend(
    title='Categoría del Producto',
    title_fontsize=10,
    fontsize=10,
    bbox_to_anchor=(1, 1),
    loc='upper left',
    frameon=False
)

# --- 6. ETIQUETAS DE DATOS ---
# Añado margen superior para que no se recorten las etiquetas
ax.margins(y=0.1)

for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.0f M',               # valores sin decimales + "M"
        padding=3,
        fontsize=10,
        color='#333333',
        fontweight='bold'
    )

plt.tight_layout()
plt.show()
```

## **3. Calificación promedio de la tienda**

### Visualización cruda

```python
df_calificacion.plot(kind='bar')
```

### **Visualización con objetivo**

```python
# --- 1. PREPARACIÓN ---
# Ya lissta

# --- 2. DISEÑO & PLOT ---
fig, ax = plt.subplots(figsize=(10, 6))

# Color Azul Acero profesional
color_pro = '#003366'

# Graficamos
# CAMBIO AQUÍ: width=0.8 hace las barras más anchas, reduciendo el espacio entre ellas.
df_calificacion.plot(kind='bar', ax=ax, width=0.8, color=color_pro, legend=False)

# --- 3. AJUSTE DE ESCALA (Zoom imperativo) ---
plt.ylim(3.9, 4.1)

# --- 4. ESTILO Y LIMPIEZA ---
# TÍTULO: Centrado (loc='center'), sin subtítulo, y más pegado a la gráfica (pad=10)
ax.set_title('Promedio de Calificación por Sucursal',
             loc='center', fontsize=20, fontweight='bold', color='#333333', pad=10)

# Ejes
ax.set_xlabel('')
plt.xticks(rotation=0, fontsize=11)

# Limpieza total de ejes y bordes
ax.get_yaxis().set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('#CCCCCC')

# --- 5. ETIQUETAS DE DATOS (Limpias) ---
for container in ax.containers:
    # CAMBIO AQUÍ: fmt='%.2f' solo muestra el número con 2 decimales, sin símbolos.
    ax.bar_label(container, fmt='%.2f', padding=5, fontsize=12, color='#333333', fontweight='bold')

plt.tight_layout()
plt.show()
```

Análisis global de los productos menos vendidos

```python
producto_worst = producto_best.sort_values()
producto_worst.head(5)
```

Analisis por tienda del producto más vendido

```python
# 1. Sacamos los nombres de los 5 productos más vendidos (tu Top Global)
top_5_nombres = tiendas_totales['Producto'].value_counts().head(5).index

# 2. Filtramos la base de datos para ver SOLO esos 5 productos
df_top5_global = tiendas_totales[tiendas_totales['Producto'].isin(top_5_nombres)]

# 3. Creamos una matriz: Productos en filas, Tiendas en columnas
# crosstab cuenta cuántas veces aparece cada combinación
ventas_top_por_tienda = pd.crosstab(df_top5_global['Producto'], df_top5_global['Tienda'])

display(ventas_top_por_tienda)
```

## **5. Envío promedio por tienda**

```python
tiendas_totales.head()
```

```python
tiendas_totales.groupby('Tienda')[['Costo de envío']].mean().round(0)
```

Visualización con objetivo

```python
# --- 1. PREPARACIÓN DE DATOS ---
# Agrupamos por Tienda, sacamos el promedio del envío y ordenamos del más barato al más caro
df_envios = tiendas_totales.groupby('Tienda')[['Costo de envío']].mean().sort_values('Costo de envío')

# --- 2. DISEÑO & PLOT ---
fig, ax = plt.subplots(figsize=(10, 6))

# Color Verde Esmeralda para denotar eficiencia (costo bajo es bueno)
color_eficiente = '#27AE60'

# Graficamos
df_envios.plot(kind='bar', ax=ax, width=0.7, color=color_eficiente, legend=False)

# --- 3. AJUSTE DE ESCALA CRÍTICO (Zoom) ---
# Si no hacemos esto, las barras se ven casi del mismo tamaño.
# Le decimos que el piso del gráfico sea 20,000 en lugar de 0.
plt.ylim(bottom=20000)

# --- 4. ESTILO Y LIMPIEZA (Estilo "Clean") ---
# Título
ax.set_title('Costo Promedio de Envío por Sucursal (Menos es mejor)',
             loc='center', fontsize=16, fontweight='bold', color='#333333', pad=20)

# Ejes
ax.set_xlabel('')
plt.xticks(rotation=0, fontsize=11)

# Limpieza total
ax.get_yaxis().set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#CCCCCC')
ax.tick_params(axis='x', length=0) # Quitar rayitas del eje X

# --- 5. ETIQUETAS DE DATOS CON FORMATO DE MONEDA ---
for container in ax.containers:
    # fmt='$ {:,.0f}': Agrega signo de pesos, coma separadora de miles, y cero decimales.
    ax.bar_label(container, fmt='$ {:,.0f}', padding=5, fontsize=11, fontweight='bold', color='#333333')

plt.tight_layout()
plt.show()
```

## 6. Análisis de Ingresos Neto

```python
tiendas_totales['Ingreso_Neto'] = tiendas_totales['Precio'] - tiendas_totales['Costo de envío']
tiendas_totales.groupby('Tienda')[['Ingreso_Neto']].sum().round(0)
```

Visualización con objetivo

```python
import matplotlib.pyplot as plt

# --- 1. PREPARACIÓN DE DATOS ---
# Agrupamos, sumamos y ordenamos
df_neto = tiendas_totales.groupby('Tienda')[['Ingreso_Neto']].sum().sort_values('Ingreso_Neto', ascending=False)

# Ajustamos la escala a Millones (Dividir entre 1,000,000)
df_neto['Ingreso_Neto'] = df_neto['Ingreso_Neto'] / 1_000_000

# --- 2. COLORIMETRÍA (Semáforo) ---
# Grises para las primeras 3, Rojo para la última
colores = ['#C0C0C0'] * (len(df_neto) - 1) + ['#D62728']

# --- 3. DISEÑO ---
fig, ax = plt.subplots(figsize=(12, 6))

# Graficamos
df_neto.plot(kind='barh', y='Ingreso_Neto', ax=ax, color=colores, width=0.65, legend=False)

# --- 4. ESTILO Y LIMPIEZA ---
# Título con el hallazgo principal
ax.set_title('Ingreso Neto (Ventas - Envíos): La Tienda 4 es la menos rentable',
             loc='left', fontsize=16, fontweight='bold', color='#333333', pad=20)

# Ejes
ax.set_xlabel('Ingreso Neto (Millones de Pesos)', fontsize=11, color='#555555')
ax.set_ylabel('')

# Limpieza de bordes (Spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#DDDDDD')
ax.spines['bottom'].set_color('#DDDDDD')

# Quitar ticks del eje Y
ax.tick_params(axis='y', length=0)

# --- 5. ETIQUETAS DE DATOS ---
for container in ax.containers:
    # Formato sin decimales + "M"
    ax.bar_label(container, fmt='%.0f M', padding=5, fontsize=10, color='#333333', fontweight='bold')

plt.tight_layout()
plt.show()
```
